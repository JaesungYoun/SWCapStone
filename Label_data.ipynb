```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
label_data = pd.read_csv("label_dataset.csv")
```

```python
label_data
```

```python
label_av = pd.DataFrame(label_data,columns=['arousal','valence'])
```

```python
label_av
```

```python
"""
HVHA(High-Valance High-Aurousal)
LVHA(Low-Valance High-Aurousal),
LVLA(Low-Valance Low-Aurousal)
HVLA(High-Valance Low-Aurousal)

"""

label_av_list = []
for i in range(len(label_data)):
    if label_data['arousal'][i] == 0:
        if label_data['valence'][i] == 0:
            label_av_list.append("LALV")
        elif label_data['valence'][i] == 1:
            label_av_list.append("LAHV")
    elif label_data['arousal'][i] == 1:
        if label_data['valence'][i] == 0:
            label_av_list.append("HALV")
        elif label_data['valence'][i] == 1:
            label_av_list.append("HAHV")
```

```python
label_av_list
```

```python
label_avdf = pd.DataFrame(label_av_list)
label_avdf.rename(columns = {0 : 'AV'},inplace =True)
```

```python
label_avdf.to_csv("label_av.csv",index = False)
```
